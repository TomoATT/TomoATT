```python
import math
class SRC:
    def __init__(self):
        self.id = 0
        self.year    = 1998
        self.month   = 1
        self.day     = 1
        self.hour    = 0
        self.minute  = 0
        self.sec     = 0
        self.lat     = 0 / math.pi * 180
        self.lon     = 0 / math.pi * 180
        self.dep     = 0
        self.mag     = 3.0
        self.name    = "s0"
    
class REC:
    def __init__(self):
        self.id = 0
        self.lat     = 0 / math.pi * 180
        self.lon     = 0 / math.pi * 180
        self.ele     = 0
        self.name    = "r0"
```

```python
import sys
sys.path.append('../utils')
from functions_for_data import cal_dis
```

```python
# generate src:
import random

seed_value = 42
random.seed(seed_value)

srcs = []
Nsrcs = 1

for i in range(1):
    src = SRC()
    src.lat     = 40.0
    src.lon     = 90.0
    src.dep     = 50.0
    src.id      = i
    src.name    = "s%d"%(i)
    srcs.append(src)

# generate rec:
recs = []
Nt = 10
Np = 10
Nrecs = Nt*Np
count = 0
for i in range(Nt):
    for j in range(Np):
        rec = REC()
        rec.lat     = 32 + (48-32)/(Nt-1) * j   
        rec.lon     = 17 + (38-17)/(Np-1) * i   
        rec.ele     = 300
        rec.id      = count
        rec.name    = "r%d"%(count)
        recs.append(rec)
        count = count + 1


```

```python
# generate src_rec_file
import os

try:
    os.mkdir("1_src_rec_files")
except:
    print("dir models exists")

fname = '1_src_rec_files/src_rec_config.dat'
doc_src_rec = open(fname,'w')

dis_thd = 300

# loop all sources
for isrc in range(len(srcs)):
    src = srcs[isrc]
    # data info
    data_info = []

    # case 2. cs (common source difference)
    phase = "P,cs"
    for i in range(len(recs)-1):
        rec1 = recs[i]
        for j in range(i+1,len(recs)):
            rec2 = recs[j]

            dis = cal_dis(rec1.lat, rec1.lon, rec2.lat, rec2.lon)
            if (dis < dis_thd):
                data_info.append('%8d %8d %6s %9.4f %9.4f %9.1f %8d %6s %9.4f %9.4f %9.1f %s %8.3f\n'%
                                 (src.id,rec1.id,rec1.name,rec1.lat,rec1.lon,rec1.ele,rec2.id,rec2.name,rec2.lat,rec2.lon,rec2.ele,phase,0.0))
    
    Ndata = len(data_info)

    # write event info 
    doc_src_rec.write('%8d %8d %2d %2d %2d %2d %5.2f %9.4f %9.4f %8.4f %5.2f %5d %s\n'%
                 (src.id,src.year,src.month,src.day,src.hour,src.minute,src.sec,src.lat,src.lon,src.dep,src.mag,Ndata,src.name))
    # write data info
    for info in data_info:
        doc_src_rec.write(info)

doc_src_rec.close()
```
