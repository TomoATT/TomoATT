```python
# notebook for plotting the effect of SIMD on the solver performance

import numpy as np
import matplotlib.pyplot as plt


# data is the fastest time among more than 10 iteration
# grid size is 100 x 100 x 100


t_local_no_simd=0.955768
t_local_avx=0.281595
t_fugaku_no_simd=5.10849
t_fugaku_avx=2.58907


# plot


fig, ax = plt.subplots()
ax.set_title('Solver performance: 1 iteration on single core')
#ax.set_xlabel('Grid size')
ax.set_ylabel('Elapsed time (best) [s]')
ax.set_xticks([0, 1])
ax.set_xticklabels(['intel core i7', 'Fujitsu A64FX'])
ax.set_ylim(0, 10)
ax.bar(0, t_local_no_simd, color='b', label='local (no SIMD)')
ax.bar(0, t_local_avx, color='r', label='local (AVX+FMA 256bits)')
ax.bar(1, t_fugaku_no_simd, color='b', label='Fugaku (no SIMD)')
ax.bar(1, t_fugaku_avx, color='r', label='Fugaku (ARM SVE+FMA 512bits)')
ax.legend()
plt.show()

```
