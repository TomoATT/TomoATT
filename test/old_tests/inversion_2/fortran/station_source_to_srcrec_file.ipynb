```python
fsta = "ega5/input/STATIONS_5.txt"
fsrc = "ega5/input/EARTHQUAKES_5.txt"

# open the files and store coordinates
with open(fsta, "r") as f:
    lines = f.readlines()

    stations = []
    for line in lines:
        ll = line.split()
        if len(ll) == 3:
            stations.append(ll) # depth, lat, lon
            
with open(fsrc, "r") as f:
    lines = f.readlines()

    sources = []
    for line in lines:
        ll = line.split()
        if len(ll) == 3:
            sources.append(ll) # depth, lat, lon
```

```python
lines = []

year_dummy = 1999
month_dummy = 1
day_dummy = 1
hour_dummy = 0
minute_dummy = 0
second_dummy = 0
mag_dummy = 0
id_dummy = 0
phase_dummy = 'P'
dist_dummy = 0
arriv_t_dummy = 0

for i_src in range(len(sources)):
    lat = float(sources[i_src][1])
    lon = float(sources[i_src][2])
    dep = float(sources[i_src][0])
    n_rec = len(stations)
    src = [i_src, year_dummy, month_dummy, day_dummy, hour_dummy, minute_dummy, second_dummy, lat, lon, dep, mag_dummy, n_rec, id_dummy]
    lines.append(src)

    for i_sta in range(len(stations)):
        lat_rec = float(stations[i_sta][1])
        lon_rec = float(stations[i_sta][2])
        elev_rec = float(stations[i_sta][0])
        st_name_dummy = 'dummy_'+str(i_sta)
        rec = [i_src, i_sta, st_name_dummy, lat_rec, lon_rec, elev_rec, phase_dummy, dist_dummy, arriv_t_dummy]

        lines.append(rec)


# write the file
fname = 'src_rec_compare_with_fortran.txt'
with open(fname, "w") as f:
    for line in lines:
        for elem in line:
            f.write('{} '.format(elem))
        f.write('\n')
```
